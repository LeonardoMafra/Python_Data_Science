```python
#DATASET DO HISTORICO DE ACOES BRASILEIRAS

df_hist = pd.read_csv('acoes_brasileiras_historico.csv')
```

```python
df_hist.info()
```

```python
df_hist.head()
```

```python
#convertendo coluna data de objeto para date

# Convertendo a coluna "Date" para datetime
df_hist["Date"] = pd.to_datetime(df_hist["Date"], errors='coerce')


import matplotlib.pyplot as plt

# Carregar o dataframe
df_hist['Volatility'] = df_hist['High'] - df_hist['Low']

# Calcular a volatilidade média por ticker
volatility_mean = df_hist.groupby('Ticker')['Volatility'].mean()

# Visualizar os resultados
volatility_mean.sort_values().plot(kind='bar', figsize=(12, 6), title="Volatilidade Média por Ticker")
plt.ylabel('Volatilidade Média')
plt.show()

# Top 5 mais voláteis
print(volatility_mean.sort_values(ascending=False).head(10))
```

```python
# Calcular o volume médio diário

df_hist['Year'] = df_hist['Date'].dt.year
average_volume = df_hist.groupby(['Year'])['Volume'].mean()

# Visualizar os resultados
average_volume.plot(kind='line', marker='o', figsize=(12, 6), title="Volume Médio Diário por Ano")
plt.ylabel('Volume Médio')
plt.xlabel('Ano')
plt.grid(True)
plt.show()
```

```python
df_hist['Date'] = df_hist['Date'].dt.tz_localize(None)


# Filtrar os dados apenas dos últimos 5 anos
start_date = pd.Timestamp.now() - pd.DateOffset(years=5)
df_5_years = df_hist[df_hist['Date'] >= start_date]

# Calcular o retorno acumulado para cada ticker
df_5_years['Daily Return'] = df_5_years.groupby('Ticker')['Close'].pct_change()
df_5_years['Cumulative Return'] = (1 + df_5_years['Daily Return']).groupby(df_5_years['Ticker']).cumprod()

# Obter o último valor do retorno acumulado para cada ticker
final_returns = df_5_years.groupby('Ticker')['Cumulative Return'].last()

# Ordenar os tickers com base no retorno acumulado e pegar os 10 maiores
top_10_growth = final_returns.sort_values(ascending=False).head(10)

# Exibir os resultados
print("As 10 ações que mais cresceram nos últimos 5 anos:")
print(top_10_growth)
```

```python
# Remover timezone das datas
df_hist['Date'] = pd.to_datetime(df_hist['Date']).dt.tz_localize(None)

# Filtrar os dados apenas dos últimos 5 anos
start_date = pd.Timestamp.now() - pd.DateOffset(years=5)
df_5_years = df_hist[df_hist['Date'] >= start_date]

# Filtrar tickers com volume médio superior a 100.000
average_volume = df_5_years.groupby('Ticker')['Volume'].mean()
valid_tickers = average_volume[average_volume > 100_000].index

# Manter apenas os tickers com volume médio acima de 100.000
df_5_years = df_5_years[df_5_years['Ticker'].isin(valid_tickers)]

# Calcular o retorno acumulado para cada ticker
df_5_years['Daily Return'] = df_5_years.groupby('Ticker')['Close'].pct_change()
df_5_years['Cumulative Return'] = (1 + df_5_years['Daily Return']).groupby(df_5_years['Ticker']).cumprod()

# Obter o último valor do retorno acumulado para cada ticker
final_returns = df_5_years.groupby('Ticker')['Cumulative Return'].last()

# Ordenar os tickers com base no retorno acumulado e pegar os 10 maiores
top_10_growth = final_returns.sort_values(ascending=False).head(10)

# Exibir os resultados
print("As 10 ações que mais cresceram nos últimos 5 anos (com volume médio > 100.000):")
print(top_10_growth)
```
