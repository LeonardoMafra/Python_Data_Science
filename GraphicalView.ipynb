```python
#configurao de estilo


sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9,5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
# CR7 X MESSI

#DADOS

ano = range(2011,2019)
messi = [59,91,45,58,52,59,54,51]
cr7 = [60,63,69,91,57,55,53,49]

# primeiro grafico de linha

plt.plot(ano,messi,'b-x',linewidth = 4, markersize = 12, markeredgewidth = 4,markeredgecolor = 'navy')


# segundo grafico de linha

plt.plot(ano,cr7,'r--o', linewidth = 4, markersize = 12,)


# titulo

plt.title(['Messi vc CR7'])


# Legenda

plt.legend(['Messi', 'CR7'])


# eixos

plt.xlabel('Ano')
plt.ylabel('Gols')
```

```python
#salvar a imagem

plt.savefig('grafico.png')
```

```python
df = pd.read_csv('Iris.csv')
```

```python
iris = df.copy()
```

```python
iris.head(5)
```

```python
# grafico de dispersao

sns.scatterplot(data = iris, x = 'SepalLengthCm', y = 'SepalWidthCm',hue = 'Species')

# titulo

plt.title('Flores')
```

```python
#Hustograma para distribuio dos dados

plt.title('Distribuio Sepal Width')
sns.distplot(iris.SepalWidthCm, kde=False)
```

```python
# Mapa de calor com o Seaborn

# carregando dataset
# reformula os dados ( produza uma tabela pivot) com base nos valores da coluna.Usa os valores exclusivos do indice especificado

voos = sns.load_dataset('flights').pivot_table( index = 'month',
                                                 columns  = 'year',
                                                 values = 'passengers')



# titulo

plt.title('Numero de Passageiros (1000)')

# desenhar o grafico


sns.heatmap(
            voos,
            fmt='.0f',
            annot = True,
            linewidths= .5,
            cmap = 'Reds',
            annot_kws= {'fontsize':10})
```

```python
# grafico de contornos - countor plot


plt.title('Flowers')

sns.kdeplot(data = iris,
            x = 'SepalLengthCm',
            y = 'SepalWidthCm',
            shade = True,
            shade_lowest = False,
            legend = True,
            hue = ''
            )
```

```python
# grafico de contornos - countor plot


plt.title('Flowers')

sns.kdeplot(data = iris,
            x = 'SepalLengthCm',
            y = 'SepalWidthCm',
            shade = True,
            shade_lowest = False,
            hue = 'Species',
            fill = True
            )
```

```python
#Boxplot ou diagrama de caixa

# dataset tips

tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
plt.title('Gorjeta total pessoa fumante x no fumante')


sns.boxplot(
    data = tips,
    x = 'day',
    y = 'total_bill',
    hue = tips.smoker
)
```

```python
# grafico  de barras com valor medio - Bar Chart

plt.title('Gorjeta total pessoa separado por sexo')


sns.barplot(
    x = 'day',
    y = 'total_bill',
    hue = 'sex',
    data = tips
)
```
