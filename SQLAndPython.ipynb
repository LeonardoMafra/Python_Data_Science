```python
#conexão python sql

import sqlite3
import pandas as pd
```

```python
#criar conexão

conn = sqlite3.connect('test.db')
print('Conexão Aberta!')
```

```python
# criando tabela

conn.execute(
    ''' CREATE TABLE IF NOT EXISTS Empregado(
      matricula integer PRIMARY KEY AUTOINCREMENT,
      nome string,
      CPF integer
    )
''')

conn.commit()

print('Tabela criada com sucesso!')
```

```python
#Populando tabelas

conn.execute("INSERT INTO Empregado VALUES(1,'Leonardo Mafra', 000);")
conn.execute("INSERT INTO Empregado VALUES(2,'Maria de Fatima', 111);")
conn.execute("INSERT INTO Empregado VALUES(3,'Fernando Silva', 222);")
conn.execute("INSERT INTO Empregado VALUES(4,'Elizabete Pinheiro', 333);")
conn.commit()
```

```python
# buscando dados da tabela

cursor = conn.execute('''
  SELECT nome,CPF
  FROM Empregado
''')

for row in cursor:
  print(row)
```

```python
#usando pandas para buscar uma query e guardar no dataframe

qry = """
  SELECT * FROM Empregado
"""

df = pd.read_sql_query(qry,conn)

df.head()
```

```python
#criar conexão

conn = sqlite3.connect('pratica.db')
print('Conexão Aberta!')
```

```python
# criar tabela e inserir dados

# criando tabela

conn.execute(
    ''' CREATE TABLE IF NOT EXISTS DF(
      CustomerID integer PRIMARY KEY,
      nome string
    )
''')

conn.commit()
```

```python
#Populando tabelas

conn.execute("INSERT INTO DF VALUES(1,'Doe');")
conn.execute("INSERT INTO DF VALUES(2,'Jo');")
conn.execute("INSERT INTO DF VALUES(3,'Tod');")

conn.commit()
```

```python
# buscando dados da tabela

cursor = conn.execute('''
  SELECT CustomerID,nome
  FROM DF
''')

for row in cursor:
  print(row)
```

```python
#usando pandas para buscar uma query e guardar no dataframe

qry = """
  SELECT * FROM DF
"""

df = pd.read_sql_query(qry,conn)

df.head()
```

```python
#usando pandas para buscar uma query e guardar no dataframe

qry = """
  SELECT * FROM DF LIMIT 1
"""

df = pd.read_sql_query(qry,conn)

df.head()
```

```python
qry = """
  SELECT * FROM DF
  WHERE CustomerID = 2
"""

df = pd.read_sql_query(qry,conn)

df.head()
```

```python
qry = """
  SELECT * FROM DF
  WHERE nome LIKE '%d'
"""

df = pd.read_sql_query(qry,conn)

df.head()
```
